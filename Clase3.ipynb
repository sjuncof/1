# Data Mungling/ Wranlgling


```python
import pandas as pd
import numpy as np
```

```python
x = {'Company':['Ford','VW','Bmw','Cooper','Cooper','Yobish'],
     'Stars':[1,2,np.nan,3,2,2],
     'Weight': [2,4,2,2,3,None],
     'Origin':['Mexico','China','China','Mexico','Uk','None',],
     'Length': [40,30,np.nan,56,pd.NaT, 44]
     }
```

```python
type(x)
```
